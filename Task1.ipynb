```python

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("population_by_country_2020-selected-columns.csv")
df.head()
print(df.columns)
df.info()
df.describe()
plt.figure(figsize=(8,5))

plt.hist(df["Med. Age"], bins=10, edgecolor="black")

plt.title("Distribution of Median Age")

plt.xlabel("Median Age")

plt.ylabel("Number of Countries")

plt.show()
```
